## Sub-step 1: Data Audit
- Load dataset
- Inspect missing values, data types, inconsistencies


```python

import pandas as pd

df = pd.read_csv('hospital_records.csv')

print(df.head())
print(df.info())
print(df.describe())

# Missing values
print(df.isnull().sum())
```


## Sub-step 2: Data Cleaning
- Handle missing values
- Fix data types
- Remove duplicates


```python

# Example cleaning

df = df.drop_duplicates()

# Fill missing values
for col in df.select_dtypes(include='number').columns:
    df[col] = df[col].fillna(df[col].median())

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Cleaned Data Shape:", df.shape)
```


## Sub-step 3: Neural Network (NumPy)
- 3-layer NN
- Forward + Backprop


```python

import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Dummy example
X = np.random.rand(100, 5)
y = np.random.randint(0, 2, (100, 1))

# Initialize weights
W1 = np.random.randn(5, 4)
W2 = np.random.randn(4, 3)
W3 = np.random.randn(3, 1)

lr = 0.01

for epoch in range(100):
    # Forward
    Z1 = sigmoid(X @ W1)
    Z2 = sigmoid(Z1 @ W2)
    y_pred = sigmoid(Z2 @ W3)

    # Loss
    loss = np.mean((y - y_pred)**2)

    # Backprop
    dW3 = Z2.T @ (y_pred - y)
    dW2 = Z1.T @ ((y_pred - y) @ W3.T)
    dW1 = X.T @ (((y_pred - y) @ W3.T) @ W2.T)

    # Update
    W3 -= lr * dW3
    W2 -= lr * dW2
    W1 -= lr * dW1

    if epoch % 10 == 0:
        print("Loss:", loss)
```


## Sub-step 4: Evaluation


```python

from sklearn.metrics import accuracy_score

preds = (y_pred > 0.5).astype(int)
print("Accuracy:", accuracy_score(y, preds))
```


## Sub-step 5: Cost-Based Decision
- Define cost of FN vs FP


```python

# Example cost
cost_fn = 10  # missing patient
cost_fp = 2   # false alarm

total_cost = ((y == 1) & (preds == 0)).sum() * cost_fn + ((y == 0) & (preds == 1)).sum() * cost_fp

print("Total Cost:", total_cost)
```
