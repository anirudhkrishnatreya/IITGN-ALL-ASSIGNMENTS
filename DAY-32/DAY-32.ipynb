# Decision Trees & Random Forest Assignments (AM + PM)
Complete Solutions

```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, recall_score
from sklearn.inspection import permutation_importance
import matplotlib.pyplot as plt
np.random.seed(42)
```

## AM Assignment - Loan Approval

```
n = 2000
data = pd.DataFrame({
    'annual_income': np.random.randint(20000, 150000, n),
    'credit_score': np.random.randint(300, 850, n),
    'loan_amount': np.random.randint(1000, 50000, n),
    'employment_years': np.random.randint(0, 20, n),
    'debt_to_income': np.random.uniform(0.1, 0.6, n),
    'num_credit_cards': np.random.randint(1, 10, n)
})

data['approved'] = ((data['credit_score'] > 650) & (data['debt_to_income'] < 0.4)).astype(int)

X = data.drop('approved', axis=1)
y = data['approved']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```
dt = DecisionTreeClassifier(max_depth=4)
dt.fit(X_train, y_train)
print(export_text(dt, feature_names=list(X.columns)))
```

```
rf = RandomForestClassifier()
params = {'n_estimators':[50,100,200],'max_depth':[3,5,10]}
search = RandomizedSearchCV(rf, params, cv=5)
search.fit(X_train, y_train)
best_rf = search.best_estimator_
```

```
def evaluate(model):
    pred = model.predict(X_test)
    return accuracy_score(y_test, pred)

print("DT Accuracy:", evaluate(dt))
print("RF Accuracy:", evaluate(best_rf))
```

## PM Assignment - Fraud Detection

```
n = 3000
data2 = pd.DataFrame({
    'claim_amount': np.random.randint(1000, 100000, n),
    'num_claims': np.random.randint(1, 10, n),
    'policy_age': np.random.randint(1, 20, n),
    'fraud_history': np.random.randint(0, 2, n)
})

data2['fraud'] = ((data2['claim_amount'] > 50000) & (data2['fraud_history']==1)).astype(int)

X2 = data2.drop('fraud', axis=1)
y2 = data2['fraud']

X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.2)
```

```
dt2 = DecisionTreeClassifier(max_depth=5)
dt2.fit(X_train2, y_train2)
print(export_text(dt2, feature_names=list(X2.columns)))
```

```
rf2 = RandomForestClassifier()
params = {'n_estimators':[100,200],'max_depth':[5,10]}
search2 = RandomizedSearchCV(rf2, params, cv=5, scoring='recall')
search2.fit(X_train2, y_train2)
best_rf2 = search2.best_estimator_

print("RF Recall:", recall_score(y_test2, best_rf2.predict(X_test2)))
```
